```python
data = data.dropna(how='any')
```

```python
def removePlotBorder():
    for spine in plt.gca().spines.values():
        spine.set_visible(False)

def displayPlotTitle(bar_plot):
    for bari in bar_plot:
        height = bari.get_height()
        plt.gca().text(bari.get_x() + bari.get_width()/2, bari.get_height() + 1, str(int(height)) + '%',
                 ha='center', color='black', fontsize=15)
        
def displayPlotTitleFloat(bar_plot,font_size):
    for bari in bar_plot:
        height = bari.get_height()
        plt.gca().text(bari.get_x() + bari.get_width()/2, bari.get_height() + 0.2, str(round(float(height),2)),
                 ha='center', color='black', fontsize=font_size)
        
def rotateTitle45(x,rotate_val):
    for item in x.get_ticklabels():
        item.set_rotation(rotate_val)
        
def floatToInt(myDF,columnName):
    return myDF[columnName].astype(int)
```

### Bölüm 4 - Veri Görselleştirme

```python
plt.figure()

#En Çok Film Çekilen Şehirler Nerelerdir?

# Şehirler sayılarına göre gruplandı.
country_info = data.groupby(['country'])['id'].count().reset_index(name='country_count')

# En yüksek ve en düşük şehir sayısı bulundu.
ct_min = country_info['country_count'].min()
ct_max = country_info['country_count'].max()

# Şehirlerin yüzdeleri alındı ve en çok film çekilen 5 şehir gözlemlendi.
country_info['country_count'] = list(map((lambda x: (x / len(data.index))*100),country_info['country_count']))
country_info = country_info.sort_values(by=['country_count'], ascending=False)
country_info = country_info[:5]

country_count_range = range(len(country_info))

# Show bar chart
country_bar = plt.bar(country_count_range, country_info['country_count'], width = 0.6, color="grey")
country_bar[0].set_color('red')
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')    
plt.xticks(country_count_range, country_info['country'])

removePlotBorder()
displayPlotTitle(country_bar)

plt.title("Filmlerin Çekildiği Ülkelerin Yüzdesi",alpha=0.8)
plt.show()
```

```python
plt.figure()

counts = []
counts.extend(country_info['country_count'])

def printChartLegend():
    liste = []
    for index,text in enumerate(country_info['country']):
        yaz = text + ' - %' + str(round(counts[index],2))
        liste.append(yaz)
    return liste

# Pie Chart label ve değerleri hazırlandı
labels= printChartLegend()
sizes = country_info['country_count']
colors = ['gold', 'yellowgreen', 'lightcoral', 'lightskyblue', 'red' ]

# Pie Chart display
plt.pie(sizes, colors=colors, startangle=90,shadow=True)
plt.legend(labels, loc=(-0.15, 0.05), shadow=True)

plt.title("Filmlerin En Çok Çekildiği Ülkelerin Yüzdesi")
plt.axis('equal')
plt.show()
```

```python
# Filmlerin IMDB skorları ile ülkeler arasında bir ilişki var mı?

# id,country ve imdb_score kolonlarını getir
imdb_ct = data[['id','country','imdb_score']]

# Ülkelerin sayısını isimlerine göre grupla
imdb_ct['ct_count'] = data.groupby(['country'])['id'].transform('count')

# Ülke sayısı 3 ten fazla olanları listeye doldur
cond = (imdb_ct['ct_count'] > 3)
condition_country = imdb_ct[cond]

imdb_mean = condition_country.groupby(['country'],as_index=False).mean()
imdb_mean['imdb_score'] = list(map((lambda x: round(x,2)),imdb_mean['imdb_score']))
imdb_mean = imdb_mean.sort_values(by=['imdb_score'], ascending=False)
```

```python
# Yorum: Film çekilen ülkelerin imdb ortalamasına bakıldığında ilk 5 sırada Brezilya,İran,Yeni Zelanda, Danimarka ve İtalya gibi
# ülkeleri görmekteyiz. Amerika,İngiltere,Fransa ve Almanya gibi en çok film çekilen ülkelerin ise imdb skorları 6.85'in altındadır.
# Çekilen film sayısı fazla olduğundan bu ülkelerin genel imdb ortalaması, film sayısı daha az olan ülkelere oranla daha düşüktür.

plt.figure()

imdbMeanPlot = range(len(imdb_mean['country']))

# Show bar chart
imdbMean_bar = plt.bar(imdbMeanPlot, imdb_mean['imdb_score'], width = 0.8, color="grey")
imdbMean_bar[0].set_color('red')
plt.tick_params(top='off', bottom='off', left='off', right='off', labelleft='off', labelbottom='on')    
plt.xticks(imdbMeanPlot, imdb_mean['country'])
plt.subplots_adjust(bottom=0.25) 

x = plt.gca().xaxis
rotateTitle45(x,90)

removePlotBorder()
displayPlotTitleFloat(imdbMean_bar,6)

plt.title("Filmlerin Çekildiği Ülkeler ve IMDB Puanları",alpha=0.8)
plt.show()
```

### Bölüm 6 - Makine Öğrenimi Uygulaması
IMDB Skor tahmini için regresyon ve filmlerin sınıflandırılması için Naive Bayes sınıflandırma yöntemini kullandım.

Cevap(Response) Değişkeni: Regresyon -> imdb_score, Classification -> imdb_class

Regresyon -> RFECV(Kullanılan), RFE(Denenen)
Classification -> Gaussian Naive Bayes(Kullanılan), AdaBoost ve RandomForest(Denenen)

```python
#Lineer Regresyon sonuçlarını iyileştirmek için object türünde olan Director_name kolonu belirli kriterler göz önünde bulundurularak
# float türünde olan Director_score alanına dönüştürüldü.

newDataFrame = data.copy()

director_likes = data[['director_name','movie_facebook_likes','imdb_score']]
director_likes['like_count'] = data.groupby(['director_name'])['id'].transform('count')
director_likes = director_likes.groupby(['director_name'],as_index=False).mean()

# Yönetmenlere Skor Ataması Yapıldı.
director_score = 0;
score_list = list()
for director in newDataFrame.values:
    for x in director_likes.values:
        if x[0].strip() == director[2].strip():
                director_score = 100 * x[2]
                break
    for best_director in director_imdb.values:
        if director[2].strip() == best_director[0].strip():
                director_score = director_score + (best_director[1] * best_director[2]) * 99
                break
    score_list.append(director_score)

# 1-100 Arası Normalizasyon Yapıldı.
max_score = max(score_list)
min_score = min(score_list)
#score_list = list(map((lambda x: round(((x-min_score) / (max_score - min_score)) * 100,2)),score_list))

score_frame = pd.DataFrame(np.array(score_list),columns=['director_score'])
```

```python
newDataFrame.reset_index(drop=True, inplace=True)
score_frame.reset_index(drop=True, inplace=True)
newDataFrame = pd.concat([newDataFrame, score_frame],axis=1)
```

### Bölüm 7 - Model Seçimi ve Değerlendirilmesi
Statsmodel kütüphanesini kullanarak regresyon modeli oluşturduğumda modeli değerlendirmek için kullandığım değerlendirme metriği:R-squared metriğidir.

Sklearn kütüphanesini kullanarak regresyon modeli oluşturduğumda modeli değerlendirmek için kullandığım değerlendirme metriği: Mean Squared Error metriğidir. Aynı zamanda kullanılan doğrulama(validation) yöntemi cross validation yöntemidir.

Regresyon modeli oluştururken feature selection metodu olarak RFECV metodunu kullandım. Aynı zamanda feature selection yöntemlerinden Recursive Feature Elimination(RFE) metodunu kullandım fakat data sayımın fazla olmasından dolayı metod sonlanmadı.

Sklearn kütüphanesini kullanarak classification modeli oluşturduğumda modeli değerlendirmek için kullandığım değerlendirme metriği:Accuracy. Aynı zamanda kullanılan doğrulama(validation) yöntemi cross validation yöntemidir.

Classification modeli oluştururken feature selection metodu olarak KBest algoritmasını kullandım. Daha iyi sonuçların elde edilip edilemeyeceğini görmek için LinearSVC ve feature selection metodlarından SelectFromModel metodunu kullandım.Fakat SelectKBest metodundan elde ettiğim accuracy değeri daha yüksek çıktı.


```python
import statsmodels.formula.api as smf 
reg = smf.ols('imdb_score ~ duration + gross + title_year + num_user_for_reviews + director_score + actor_3_facebook_likes + num_critic_for_reviews + num_voted_users', data=newDataFrame).fit()
print (reg.summary())

# Regresyon Modeli : duration + gross + title_year + num_user_for_reviews + director_score + actor_3_facebook_likes + num_critic_for_reviews + num_voted_users'
# 8 kolondan oluşmaktadır. Çıkan sonuçları yorumlamak istersek bu modeldeki değerlendirme metriğim R-squared değeri.
# Çıkan R-squared değerinin açıklaması şu şekildedir: 8 özellikten oluşan bu model bütün datadaki değişikliğin yüzde 35.8'ini
# açıklamaktadır.
```

### RFECV Öncesi

```python
# Sklearn kütüphanesini kullanarak regresyon modelini oluşturmak için gerekli olan optimal özellik sayısını belirliyorum.
# Bu metod sırayla çıkan grid_score değerlerine göre değişken sayılarını arttıyor ve optimal özellik sayısını bulmayı sağlıyor.
# Sonuçları okuduğumuzda 8 özellik ile en iyi mean-squared-erroru yakalayabileceğimiz söyleniyor. Bu skoru yakalamak için
# hangi kolonları kullanmamız gerektiğide support_ dizisinde belirtiliyor. Support dizisine göre 2. özelliği yani budget özelliğini
# modelden çıkarmalıyız.
from sklearn import linear_model
from sklearn.feature_selection import RFECV
regr = linear_model.LinearRegression()

X = newDataFrame[['duration','budget','facenumber_in_poster','director_score','gross','title_year','num_user_for_reviews','num_critic_for_reviews','num_voted_users']]
y = newDataFrame['imdb_score']

rfecv = RFECV(estimator=regr, cv=5, scoring='mean_squared_error')
rfecv.fit(X, y)
print("Optimal number of features : %d" % rfecv.n_features_)
print(rfecv.support_)
print(rfecv.grid_scores_)

plt.figure()
plt.xlabel("Number of features selected")
plt.ylabel("Cross validation score")
plt.plot(range(1, len(rfecv.grid_scores_) + 1), rfecv.grid_scores_)
plt.show()
```

### RFECV Sonrası

```python
# Yukarıda bulduğum özellikleri kullanarak bir model oluşturduğumda 0.32 r-squared ve -0.74 mean squared error değerlerini gözlemledim.
# Feature selection yapmadan modelimi kullandığımda 0.29 r-squared ve -0.77 mean squared error değerlerini gördüm.
# Regresyon modelimde çok büyük bir artış gözlemleyemedim fakat feature selection kullanarak oluşturduğum modelinin değerlerine
# başka yöntemlerle ulaşamadım.

from sklearn.model_selection import cross_val_score
regr1 = linear_model.LinearRegression()

X = newDataFrame[['duration','facenumber_in_poster','director_score','gross','title_year','num_user_for_reviews','num_critic_for_reviews','num_voted_users']]
y = newDataFrame['imdb_score']

scoresMSE= cross_val_score(regr1, X, y, cv=10, scoring='mean_squared_error') 
print("Mean Squared Errors: "+ str(list(scoresMSE)))
print('Mean:%.2f' %scoresMSE.mean())

scoresR2= cross_val_score(regr1, X, y, cv=10, scoring='r2') 
print("R2 scores: "+ str(list(scoresR2)))
print("r2_score: %.2f" % scoresR2.mean())
```

### SelectKBest ve Cross Validation Sonrası

```python
train_columns = testData[['num_voted_users','duration','director_score','cast_total_facebook_likes']] # Accuracy = 0.79
# train_columns = testData[['num_voted_users','cast_total_facebook_likes']] -> Accuracy = 0.80
target_columns = testData['imdb_class']
X = np.array(train_columns)  
y = np.array(target_columns)
```

```python
# KBest algoritması kullanmadan lineer regresyon modelinde kullandığım modeli kullanarak 0.33 accuracy değerini elde ettim.
# İlk olarak GaussianNB Sınıflandırma algoritmasını kullandım.
# Daha sonra sınıflandırma modelini KBest algoritmasının verdiği sonuçlara göre oluşturdum. Ardından ilk olarak cross validation
# kullanmadan sürekli aynı test ve train parçalarını kullanarak modelimin accuracy değerine baktım. Fakat bu değer bizi
# her zaman doğru sonuca götürmeyebilir. Çünkü random_state parametresi değiştiğinde accuracy değeride değişmiş olacak.
# Bu yüzden daha doğru sonuçlar elde etmek için cross validation yöntemini kullandım.

from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import train_test_split
from sklearn.model_selection import cross_val_score

X_train, X_test, y_train, y_test= train_test_split(X, y, test_size= 0.3, random_state=11) # 11 = %79, 15 = %67, 40 = % 78, 400 = %77

clf = GaussianNB()
clf.fit(X_train, y_train) 

clf.score(X_test,y_test)
```

### LinearSVC & SelectFromModel

```python
from sklearn.svm import LinearSVC
from sklearn.feature_selection import SelectFromModel

lsvc = LinearSVC(C=0.05, penalty="l1", dual=False).fit(X, y)
model = SelectFromModel(lsvc, prefit=True)
X_new = model.transform(X)

scoresAcc= cross_val_score(clf, X_new, y, cv=5, scoring='accuracy')

print("LinearSVC Accuracy values: "+ str(list(scoresAcc)))
print('LinearSVC Mean:%.2f' %scoresAcc.mean())
```

### SelectKBest & Chi2

```python
# Cross validation yöntemini kullanarak verdiğim cv=10 parametresine göre data 10 test parçasına ayrıldı. Yani her seferde datanın
# bir parçası test olarak geriye kalan 9 parçası training datası olarak kullanılacak. Böylece datanın bütün parçaları test için
# kullanılmış oldu. Bu parçaların ortalaması alınarak accuracy değeri belirlendi. Bu sonuç araştırmam için daha güvenilir ve
# etkili bir sonuçtur. Aynı zamanda KBest ve chi2 yöntemlerini kullanarak en başta modelimin en önemli 4 özelliğinin seçilmesini
# istedim, çıkan accuracy değeri 0.77 oldu. K parametresini 2 yaptığımda yine 0.77 accuracy değerine ulaştım. Yani modelimin
# sadece 2 özelliği ile yüzde 77 doğrulukla sınıflandırma yapabilirim. Peki 0.77 accuracy değerini veren özelliklere nasıl ulaştım?
# KBest algoritması çalıştırıldıktan sonra scores_ değişkeni ile verdiğim modeldeki özelliklerin skorlarını görüntüledim.
# Bu sonuçlara göre num_voted_users ve cast_total_facebook_likes kolonları 0.77 accuracy değerini veren kolonlar oldu.

from sklearn.feature_selection import SelectKBest
from sklearn.feature_selection import chi2
feat_extractor = SelectKBest(chi2, k=2)
X_new = feat_extractor.fit_transform(X, y)
scoresAcc= cross_val_score(clf, X_new, y, cv=10, scoring='accuracy')

print("SelectKBest Columns Scores:" + str(feat_extractor.scores_))
print("SelectKBest Accuracy values: "+ str(list(scoresAcc)))
print('SelectKBest Mean:%.2f' %scoresAcc.mean())
```
